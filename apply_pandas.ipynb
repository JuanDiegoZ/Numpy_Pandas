__Apply__
Apply Es un comando muy poderoso que nos deja aplicar funciones a nuestro DataFrame

```python
import pandas as pd
df_ciclyng = pd.read_csv('/home/jann/Platzi/DataScience/CourseNumpyAndPandasBasic/stages_TDF (1).csv', sep =',')


def two_times(value):
    return value  + 2
```

```python
df_ciclyng['Distance x2'] = df_ciclyng['Distance'].apply(two_times)
df_ciclyng
```

```python
df_ciclyng['Distance x3'] = df_ciclyng['Distance'].apply(lambda x:  x* 3)
df_ciclyng
```

Apply en varias columnas con condiciones, hay que especificar a que los vamos a aplicar (filas o columnas)

```python
df_ciclyng['Lambda']= df_ciclyng.apply(lambda x: x['Distance'] * 2 if x['Winner_Country'] == 'GBR' else x['Distance'], axis = 1)
df_ciclyng
```
